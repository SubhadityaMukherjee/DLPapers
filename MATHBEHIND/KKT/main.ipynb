# SLSQP -> Sequential Linear Squares Problem

```python
import numpy as np
from scipy.optimize import minimize
```

- Optimize
- nnls

```python
x = [1,5,5,1]
print(x[0:3])
np.prod(x)
```

```python
def objective(x):
    return x[0]*x[3]*(np.sum(x[0:3]))+x[2]
```

```python
def constraint1(x):
    return np.prod(x)-25.0
```

```python
def constraint2(x):
    sum_eq = 40.0
    for i in range(4):
        sum_eq = sum_eq - x[i]**2
    return sum_eq
```

```python
# initial guesses
x = np.array([1,5,5,1]);x
```

```python
# show initial objective
print('Initial SSE Objective: ' + str(objective(x)))
```

```python
bnds = tuple([(1.0,5.0) for i in range(4)]);bnds
```

```python
con1 = {'type': 'ineq', 'fun': constraint1}
con2 = {'type': 'eq', 'fun': constraint2}
con1,con2
```

```python
cons = ([con1,con2]);cons
```

```python
solution = minimize(objective,x,method='SLSQP',bounds=bnds,constraints=cons)
solution.x
```

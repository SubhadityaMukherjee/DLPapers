# Resnet

```python
class Model(nn.Module):
    def __init__(self):
        super(Model, self).__init__()
        self.conv1 = nn.Conv2d(3, 64, kernel_size=7)
        self.conv2 = nn.Conv2d(64, 128, kernel_size=3)
        self.conv3 = nn.Conv2d(128, 256, kernel_size=3)
        self.conv4 = nn.Conv2d(256, 512, kernel_size=3)
        self.rel = nn.ReLU(inplace=True)
        self.dro = nn.Dropout(0.25)
        self.res1 = res_block(64)
        self.res2 = res_block(128)
        self.res3 = res_block(256)
        self.res4 = res_block(512)
        self.ap = nn.AvgPool2d(2)
        self.mp = nn.MaxPool2d(1)
        self.lin = nn.Linear(204800, 100)

    def forward(self, x):
        x = self.conv1(x)
#         x = self.rel(x)
#         x = self.dro(x)
        x = self.mp(x)
        x = self.res1(x)
        x = self.res1(x)
        x = self.mp(x)
#         x = self.res1(x)
        x = self.conv2(x)
        x = self.res2(x)
        x = self.res2(x)
        x = self.mp(x)
#         x = self.res2(x)
#         x = self.res2(x)
        x = self.conv3(x)
        x = self.res3(x)
        x = self.res3(x)
        x = self.mp(x)
#         x = self.res3(x)
#         x = self.res3(x)
#         x = self.res3(x)
#         x = self.res3(x)
        x = self.conv4(x)
        x = self.res4(x)
        x = self.res4(x)
#         x = self.res4(x)
        x = self.mp(x)
        x = torch.flatten(x,1)
        x = self.lin(x)

        return x
```

```python
learn = None
gc.collect()
```

```python
learn = Learner(
    data,
    Model(),
    metrics=[accuracy, error_rate],
    opt_func=AdamW,
    callback_fns=ShowGraph,
).to_fp16()
```

```python
learn.lr_find()
learn.recorder.plot()
```

```python
learn.fit_one_cycle(5, 3e-6)
```

```python
learn.unfreeze()
learn.fit_one_cycle(15, 3e-5)
```

```python
learn.show_results()
```

```python
dummy_input = Variable(torch.randn(1, 3, 32, 32)).cuda().half()
torch.onnx.export(learn.model, dummy_input, "model.onnx")
netron.start("model.onnx")
```
